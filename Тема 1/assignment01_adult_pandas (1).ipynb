**7. Выведите статистику возраста для каждой расы (признак *race*) и каждого пола. Используйте *groupby* и *describe*. Найдите таким образом максимальный возраст мужчин расы *Amer-Indian-Eskimo*.**

**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак *marital-status*)? Женатыми считаем тех, у кого *marital-status* начинается с *Married* (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**

```python
married = data['marital-status'].str.startswith('Married')
rich_married_ratio = data[(data['sex'] == 'Male') & married & (data['salary'] == '>50K')].shape[0] / data[(data['sex'] == 'Male') & married].shape[0]
rich_unmarried_ratio = data[(data['sex'] == 'Male') & ~married & (data['salary'] == '>50K')].shape[0] / data[(data['sex'] == 'Male') & ~married].shape[0]
print("Доля зарабатывающих >50K среди женатых мужчин:", rich_married_ratio)
print("Доля зарабатывающих >50K среди холостых мужчин:", rich_unmarried_ratio)
```

**9. Какое максимальное число часов человек работает в неделю (признак *hours-per-week*)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_hours = data['hours-per-week'].max()
num_people_max_hours = data[data['hours-per-week'] == max_hours].shape[0]
rich_max_hours_ratio = data[(data['hours-per-week'] == max_hours) & (data['salary'] == '>50K')].shape[0] / num_people_max_hours
print("Максимальное количество рабочих часов в неделю:", max_hours)
print("Число людей, работающих столько часов:", num_people_max_hours)
print("Процент зарабатывающих >50K среди них:", rich_max_hours_ratio)
```
